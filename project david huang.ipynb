```python
import requests
from bs4 import BeautifulSoup
response = requests.get('https://money.cnn.com/data/sectors/tech/communications/?sector=4900&industry=4910&page=1')
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
rows = soup.find_all('td')
tickers=[]
for row in rows:
    cells = row.find('a')
    if cells:
        for cell in cells:
            title = cell
            tickers.append(title)
#tickers.remove(tickers[0])
#tickers.remove(tickers[1])
#tickers.remove(tickers[2])
#tickers.remove(tickers[3])
#tickers.remove(tickers[4])
#tickers.remove(tickers[5])
print(tickers)
print(len(tickers))
```

```python
market_capital = []
tr = soup.find_all('tr')
for marketc in tr:
    marketcap = marketc.find('td',"wsod_aRight")
    if marketcap:
        for money in marketcap:
            moneybag = money
            market_capital.append(moneybag)
market_capital_adjust = []
for adjust in market_capital:
    if len(adjust)>1:
        market_capital_adjust.append(adjust)
print(market_capital_adjust)
print(len(market_capital_adjust))
```
